# Глава 7. Как изобразить нейронную сеть

| На языке алгебры | Перевод |
| ----------- | ----------- |
| $l$ | Слой |
| $W$ | Веса |
| $l_0W_0$ | Умножить вектор слоя $l_0$ на весовую матрицу $W_0$ |
| $l_1W_1$ | Умножить вектор слоя $l_1$ на весовую матрицу $W_1$ |
| $l_1 = relu(l_0W_0)$ | Чтобы получить вектор слоя $l_1$, нужно умножить вектор слоя $l_0$ на весовую матрицу $W_0$ и пропустить результат через функцию активации $relu$ (преобразует отрицательные числа в 0) |
| $l_2 = l_1W_1$ | Чтобы получить вектор слоя $l_2$, нужно умножить вектор слоя $l_1$ на весовую матрицу $W_1$ |
| $l_2 = relu(l_0W_0)W_1$ | То есть всю логику этапа прямого распространения можно выразить единственной формулой. Эта формула предполагает, что все матрицы и векторы имеют совместимые размеры |

## Сравнение разных способов визуализации

`layer_2 = relu(layer_0.dot(weights_0_1)).dot(weights_1_2)`
