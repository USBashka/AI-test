# Глава 2. Основные понятия

Глубокое обучение — это подмножество методов машинного обучения, области изучения и создания машин, способных обучаться

Машинное обучение — это область исследований, которая даёт компьютерам способность обучаться без непосредственного программирования

__Машинное обучение ~= Обезьяна видит, обезьяна делает__

## Обучение с учителем и без

Машинное обучение с учителем — это прямая имитация закономерностей, имеющих место между двумя наборами данных. В нём всегда __входной__ набор данных преобразуется в _выходной_. Например:
- Использование __пикселей__ изображение для определения _присутствия или отсутствия кота_
- Использование __списка понравившехся фильмов__ для выбора _фильмов, которые могут понравиться_
- Использование __метеорологических данных__ для предсказания _погоды_

Что известно -> __Обучение с учителем__ -> Что хотелось бы знать

Обучение без учителя группирует данные. Набор данных, в который происходит преобразование, прежде не был известен.

## Параметрическое и непараметрическое обучение

Параметричность задаёт способ _хранения_ результатов и зачастую _метод обучения_

Параметрическая модель характеризуется наличием _фиксированного_ числа параметров, тогда как непараметрическая модель имеет _бесконечное_ число параметров (определяется данными)

Глубокое обучение является классом параметрических моделей
